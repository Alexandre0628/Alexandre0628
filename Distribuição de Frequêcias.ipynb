```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
enem_sp = pd.read_csv(r"C:\Users\Theki\OneDrive\Área de Trabalho\analise de dados\enem_2019.csv", sep=',', encoding='iso-8859-1')
```

```python
enem_sp.head()
```

```python
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
```

```python
enem_sorocaba.head()
```

```python
enem_sorocaba.shape
```

```python
idades = enem_sorocaba['IDADE']
```

```python
idades
```

```python
freq_absoluta = Counter(idades)
```

```python
freq_absoluta
```

```python
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
```

```python
freq_absoluta
```

```python
freq_absoluta = freq_absoluta.sort_index(ascending=True)
freq_absoluta
```

```python
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

```python
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc
```
